```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import preprocessing
#from sklearn.tree._export import plot_tree
```

```python
labels = ["age","income","student","credit_rating","buys_computer"]
df = pd.read_csv("cpu.csv",names=labels)
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
df
```

```python
train_features = df.iloc[:8,:-1]
test_features = df.iloc[8:,:-1]
train_targets = df.iloc[:8,-1]
test_targets = df.iloc[8:,-1]
tree = DecisionTreeClassifier(criterion='gini').fit(train_features,train_targets)
prediction = tree.predict(test_features)
for name, importance in zip(test_features.columns, tree.feature_importances_):
    print(name, importance)
print("The prediction accuracy is: ",
      tree.score(test_features,test_targets)*100,"%")
```
