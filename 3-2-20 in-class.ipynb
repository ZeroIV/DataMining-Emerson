```python
train_df['Age'].fillna(train_df.groupby('Pclass')['Age'].transform("mean"),inplace=True)
```

```python
train_df.head()
```

```python
sns.heatmap(train_df.isnull(),yticklabels=False,cbar=False)
```

```python
train_df.drop('Cabin',axis=1 ,inplace=True)
```

```python
train_df.info()
```

```python
gender=pd.get_dummies(train_df['Sex'],drop_first=True)
embark=pd.get_dummies(train_df['Embarked'],drop_first=True)
gender.head()
embark.head()
```

```python
train_df.head()
```

```python
train_df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
train_df.head()
```

```python
train= pd.concat([train_df,gender,embark],axis=1)
```

```python
train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#Train_test_split is discussed in slide deck
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1),train['Survived'],
                                                    test_size=0.30,random_state=101)
logmodel = LogisticRegression(solver='lbfgs',max_iter=5150)
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
predictions
```
